12. percentage of chat on which they have received a feedback 

```python
agent_data = df.groupby('Agent Name')[['Total Chats', 'Total Feedback']].sum().reset_index()
agent_data['percentage'] = (agent_data['Total Feedback'] / agent_data['Total Chats'])*100
agent_data
```

13. Total contributation hour for each and every agents weekly basis 

```python
df['Average Response Time'] = pd.to_timedelta(df['Average Response Time'])

weekly_contribution = df.groupby([pd.Grouper(freq = 'W') , 'Agent Name'])['Average Response Time'].sum().reset_index()

weekly_contribution['Total Hours'] = weekly_contribution['Average Response Time'].dt.total_seconds() / 3600

weekly_contribution
```

14. total percentage of active hour for a month 

```python
df['Average Response Time'] = pd.to_timedelta(df['Average Response Time'])
monthly_contribution = df.groupby('Agent Name')['Average Response Time'].sum().reset_index()
monthly_contribution['Total Active Hours'] = monthly_contribution['Average Response Time'].dt.total_seconds()/3600

total_possible_hours = 22 * 8

monthly_contribution['Active Percentage'] = (monthly_contribution['Total Active Hours'] / total_possible_hours) * 100

monthly_contribution
```
